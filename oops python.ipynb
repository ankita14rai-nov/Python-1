```python
#instance methods => instance methods is a method or a function that is defined inside the class which is bound to/associated with the instances/object.

class Student:
    "this is a class Student to manage student information and other activities."
    def study(self, n_hours):
        print(f"self is: {self}")
        print(f"the student studies for {n_hours} hours.")

s1= Student()
print(f" The object: {s1}")

s1.study(2)
print("***************")
s2= Student()
print(f" The object: {s2}")
s2.study(3)

#when we call instance method using the object/instance of the class, the object is automatically passed as first argument to the method by python.
# that first argument is by standard convention is named as self. we can use any name but it is recommended to use self.
```

```python
class Student:
    def __init__(self, name, roll, dept):
        print(f"calling the initializer for {self}")
        self.name= name
        self.roll= roll
        self.department= dept
    def study(self, n_hours):
        print(f"the student studies for {n_hours} hours.")
    def sports(self, sports_name):
        print(f"the student plays {sports_name}.")

student1= Student("Ankita:", 123044012, dept= "CSE")
student2= Student("Sanchit:", 123044013, dept= "CSE")

student1.grade= "A"


print(student1.name, student1.roll, student1.department)
print(student2.name, student2.roll, student2.department)

print(student1.__dict__)  #returns the dictionary of the object which contains the attributes and their values.
print(student2.__dict__)


#instance variables or attributes  are different for different objects of the same class. they are created inside the __init__ method using self keyword. they are accessed using the object name and dot operator.
```

```python
#Class Variables or attributes => class variables are the variables that are shared same copy of the class among all the objects. they are created inside the class but outside the __init__ method. they are accessed using the class name and dot operator.

class Dog:
    species = "Canine"  # This is a class variable

    def __init__(self, name):
        self.name = name  # This is an instance variable

# Accessing the class variable
print(Dog.species)  # Output: Canine

d1 = Dog("Fido")
d2 = Dog("Buddy")

# Both instances share the same class variable
print(d1.species)   # Output: Canine
print(d2.species)   # Output: Canine

    
```

```python
#classmethods => class methods are the methods that are bound to the class and not the object. they can access and modify the class state that applies across all instances of the class. they are defined using @classmethod decorator and take cls as the first parameter which refers to the class itself.
class welcome:
    @classmethod
    def greet(cls):
        print(cls)
        print("hello")

w1= welcome()
w1.greet()
print("***************")
print(welcome)
```

```python
class Student:
    college_name="ABC COLLEGE"
    department= ["CSE", "ARTS", "TECH"]

    def __init__(self, name, roll, dept):
        print(f"calling the initializer for {self}")
        self.name= name
        self.roll= roll
        self.department= dept
    def study(self, n_hours):
        print(f"the student studies for {n_hours} hours.")
    def sports(self, sports_name):
        print(f"the student plays {sports_name}.")

    @classmethod
    def greet(cls):
        print(cls)
        print(f"welcome to {cls.college_name}")  

    @classmethod
    def get_departments(cls):
        print(f"the departments available are: {cls.department}")
        for department in cls.department:
            print(department)

student1= Student("Ankita:", 123044012, dept= "CSE")
student1.greet()
student1.study(2)
student1.sports("football")
student1.get_departments()

    
```

```python
#static methods => static methods are the methods that are neither bound to the class nor the object. they are defined using @staticmethod decorator and do not take any default parameter like self or cls. they are used to perform some utility function that does not access or modify the class state or instance state. they can be called using the class name or the object name. 

class Student:
    '''A class to represent a student in the college.'''
    college_name="ABC COLLEGE"
    department= ["CSE", "ARTS", "TECH"]

    def __init__(self, name, roll, dept):
        print(f"calling the initializer for {self}")
        self.name= name
        self.roll= roll
        self.department= dept
    def study(self, n_hours):
        print(f"the student studies for {n_hours} hours.")
    def sports(self, sports_name):
        print(f"the student plays {sports_name}.")

    @classmethod
    def greet(cls):
        print(cls)
        print(f"welcome to {cls.college_name}")  

    @classmethod
    def get_departments(cls):
        print(f"the departments available are: {cls.department}")
        for department in cls.department:
            print(department)

    @staticmethod
    def info():
        print("this is a student class to manage student information and other activities.")    
        

student1= Student("Ankita:", 123044012, dept= "CSE")
student1.greet()
student1.study(2)
student1.sports("football")
student1.get_departments()
student1.info()  #we can call static method using the object name.


# student2= Student("priya:", 123044016, dept= "CE")
# student2.greet()
# student2.study(3)   

```
