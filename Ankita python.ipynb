```python
def second_largest(lst):
    first = second = float('-inf')
    for num in lst:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
    return second


print(second_largest([10, 20, 4, 45, 99]))  
```

```python
#Find frequency of each element in a list.
def count_frequency(lst):
    freq = {}
    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    return freq

# Example
print(count_frequency([1, 2, 2, 3, 1, 4, 2]))
```

```python
def fabonacci(n):
    """This function returns the Fibonacci sequence up to the nth number."""
    sequence = []
    a, b = 0, 1
    for i in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence
print(fabonacci(10))
```
