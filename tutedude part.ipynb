```python
#Exercise- lists and loops
countries= ["India", "Australia", "Nepal", "Sri lanka", "Japan", "Iceland", "Iran"]

#countries started with I or not.

counter=0
for country in countries:
    if country[0]== "I":
        counter= counter+1

print(counter)
        
```

```python
import random

num=1
print("welcome to the number guessing game. We have a number that needs to be guessed. You have 10 chances. ")
print("the secret number is between 1 and 50")

secret_number= random.randint(1,50)
attempts= 10
attempts_exhausted= False

while num <=10:
    print(f"you have {attempts} attempts remaining.")
    user_input= int(input("enter your guessing number: "))
    if user_input== secret_number:
        print("congrats! your guess is right, end the game.")
        attempts_exhausted= True
        break
    else:
        if user_input < secret_number:
            higher_or_lower= "higher"
        else:
           higher_or_lower= "lower" 
        print(f"your guess is wrong! Try {higher_or_lower} number.")
    num += 1
    attempts -=1

if attempts_exhausted == False:
    print("bad luck! You couldnot guess the number")

print("bad luck! the game is over.")
    

   
```
